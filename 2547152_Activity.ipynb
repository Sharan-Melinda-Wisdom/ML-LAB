1. Central tendency of data (Visual; Numeric)

```python
mean_age=df['Age'].mean()
print("Mean Age= ",mean_age)
sns.histplot(df['Age'], bins=10, kde=True)
plt.axvline(mean_age, linestyle='--', label='Mean')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.legend()

plt.show()
```

**Interpretation:** The  

```python
print("Median Purchase Amount =",df['Purchase Amount (USD)'].median())
sns.violinplot(x=df['Purchase Amount (USD)'], color='lightgreen', inner='box')
plt.title("Purchase Amount Distribution")
plt.show()
```

```python

print("Mode Payment Method =", df['Payment Method'].mode()[0])
sns.countplot(x='Payment Method', data=df, hue='Payment Method')
plt.title("Payment Method Frequency")
plt.show()
```

2. Outlier identification (Visual ; Numeric). Add outliers if the dataset does not have any.

```python
Q1 = df['Purchase Amount (USD)'].quantile(0.25)
Q3 = df['Purchase Amount (USD)'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

outliers = df[
    (df['Purchase Amount (USD)'] < lower) |
    (df['Purchase Amount (USD)'] > upper)
]

print("Number of Outliers =", len(outliers))
print(outliers)
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(x=df['Purchase Amount (USD)'])
plt.title("Outlier Detection using Boxplot")
plt.show()
```

3. Visualizing relationship between 2 numeric variables


```python
plt.hexbin(df['Age'],df['Purchase Amount (USD)'],gridsize=15)
plt.colorbar(label='Count')
plt.title("Age vs Purchase Amount")
plt.xlabel("Age")
plt.ylabel("Purchase Amount (USD)")
plt.show()
```

4. Visualizing relationship between numeric and nominal variable

```python
sns.barplot(data=df,x='Category',y='Purchase Amount (USD)')
plt.title("Average Purchase Amount by Category")
plt.xticks(rotation=20)
plt.show()
```

5. Visualizing relationship between two nominal variables.

```python
avg_purchase = df.groupby('Age')['Purchase Amount (USD)'].mean()

plt.figure(figsize=(8,5))

plt.plot(avg_purchase.index, avg_purchase.values)

plt.title("Average Purchase Amount by Age")
plt.xlabel("Age")
plt.ylabel("Average Purchase Amount (USD)")
plt.show()
```
